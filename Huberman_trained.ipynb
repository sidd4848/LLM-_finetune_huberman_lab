# Importing model and packages

```python
import torch
from youtube_transcript_api import YouTubeTranscriptApi
import pandas as pd
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    TrainingArguments,
    pipeline,
    logging,
)
from peft import LoraConfig
from trl import SFTTrainer
from datasets import Dataset

model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

pipeline_m = pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)

messages = [
    {"role": "system", "content": "You are inspired with huberman lab podcasts"},
    {"role": "user", "content": "How to improve skin care and what routine I should follow?"},
]

outputs = pipeline_m(
    messages,
    max_new_tokens=256,
)
print(outputs[0]["generated_text"][-1])
```

# Importing Training data

```python
transcript_text = ''
# Huberman selected podcasts
video_id = ['wgUjIRtote8','CD0bRU1e1ZM','pZX8ikmWvEU',
            #'F54qXuTpgfM','79p1X_7rAMo', 'nDLb8_wgX50',
            'QmOF0crdyRU']
for i in video_id:
  transcript = YouTubeTranscriptApi.get_transcript(i)
  transcript_text = transcript_text + ' '.join([entry['text'] for entry in transcript]).replace('[Music]','')
  print(len(transcript_text))
```

```python
# Convert to dataset format
dataset = Dataset.from_dict({"text": [transcript_text]})
dataset
```

# Getting QLoRA for 4 bit computation

```python
compute_dtype = getattr(torch, "float16")

quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=compute_dtype,
    bnb_4bit_use_double_quant=False,
)
```

```python
base_model = model_id
model = AutoModelForCausalLM.from_pretrained(
    base_model,
    quantization_config=quant_config,
    device_map={"": 0}
)
model.config.use_cache = False
model.config.pretraining_tp = 1
```

```python
tokenizer = AutoTokenizer.from_pretrained(base_model, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"
```

```python
peft_params = LoraConfig(
    lora_alpha=16,
    lora_dropout=0.1,
    r=64,
    bias="none",
    task_type="CAUSAL_LM",
)
```

```python
training_params = TrainingArguments(
    output_dir="./results",
    num_train_epochs=1,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=1,
    optim="paged_adamw_32bit",
    save_steps=25,
    logging_steps=25,
    learning_rate=2e-4,
    weight_decay=0.001,
    fp16=False,
    bf16=False,
    max_grad_norm=0.3,
    max_steps=-1,
    warmup_ratio=0.03,
    group_by_length=True,
    lr_scheduler_type="constant",
    report_to="tensorboard"
)
```

```python
trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    peft_config=peft_params,
    dataset_text_field="text",
    max_seq_length=None,
    tokenizer=tokenizer,
    args=training_params,
    packing=False,
)
```

```python
new_model = "llama-2-7b-sft"
trainer.model.save_pretrained(new_model)
trainer.tokenizer.save_pretrained(new_model)
```

```python
logging.set_verbosity(logging.CRITICAL)

prompt = "How to improve skin care and what routine I should follow?"
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=200)
result = pipe(f"<s>[INST] {prompt} [/INST]")
print(result[0]['generated_text'])
```
